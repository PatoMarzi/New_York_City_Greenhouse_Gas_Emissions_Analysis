# 1.  Entender la Relación entre las Variables:
Para entender la relación entre las variables, observa las estadísticas descriptivas y visualizaciones: 


```python


# DataFrame de contaminación del aire
contaminacion = df_energy[['Country', 'Year', 'CO2_emission']]

# Filtra las emisiones de CO2 solo para Estados Unidos y el año más reciente
contaminacion_usa_latest = contaminacion[(contaminacion['Country'] == 'United States') & (contaminacion['Year'] == contaminacion['Year'].max())]

# Visualización de emisiones de CO2 para Estados Unidos en el año más reciente
sns.barplot(x='Year', y='CO2_emission', data=contaminacion_usa_latest)
plt.title('Emisiones de CO2 en Estados Unidos (Año más reciente)')
plt.show()

```

# Exploración descriptiva y calcular la correlación entre la cantidad de viajes y las emisiones de CO2

```python


# DataFrame de viajes
viajes = df_yellow[['passenger_count', 'trip_distance', 'total_amount']]

# DataFrame de contaminación del aire
contaminacion = df_energy[['Country', 'Year', 'CO2_emission']].reset_index(drop=True)

# Unir los DataFrames
df_combined = pd.concat([viajes, contaminacion], axis=1)

# Excluir la columna 'Country' al escalar los datos
df_combined_scaled = (df_combined.drop('Country', axis=1) - df_combined.drop('Country', axis=1).min()) / (df_combined.drop('Country', axis=1).max() - df_combined.drop('Country', axis=1).min())

# Visualización de la correlación (Escalado)
sns.scatterplot(x='trip_distance', y='CO2_emission', data=df_combined_scaled)
plt.title('Correlación entre la cantidad de viajes y emisiones de CO2 (Escalado)')
plt.xlabel('Cantidad de Viajes (distancia)')
plt.ylabel('Emisiones de CO2')
plt.show()



# Visualización de la correlación con rango limitado
x_range = (0, 50)  # Ajusta según tus necesidades
y_range = (0, 100)  # Ajusta según tus necesidades
sns.scatterplot(x='trip_distance', y='CO2_emission', data=df_combined)
plt.title('Correlación entre la cantidad de viajes y emisiones de CO2')
plt.xlabel('Cantidad de Viajes (distancia)')
plt.ylabel('Emisiones de CO2')
plt.xlim(x_range)
plt.ylim(y_range)
plt.show()

# Visualización de la correlación con escala logarítmica (opcional)
sns.scatterplot(x='trip_distance', y='CO2_emission', data=df_combined)
plt.title('Correlación entre la cantidad de viajes y emisiones de CO2')

plt.ylabel('Emisiones de CO2')
plt.xscale('log')  
plt.show()




```
